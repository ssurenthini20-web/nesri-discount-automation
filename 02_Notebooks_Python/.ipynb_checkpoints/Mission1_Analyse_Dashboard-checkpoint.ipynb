```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

df = pd.read_excel("/Users/surenthinisivakumar/Documents/Stage_Nesri_Discount/03_Donnees_propres/commandes_propres_v3.xlsx")
print(f"Nombre de lignes : {len(df)}")
df.head()
```

```python
df_rembourse = df[df["Refunded Amount"] > 0]

print("=" * 50)
print("ARGENT PERDU EN REMBOURSEMENTS")
print("=" * 50)
print(f"Montant total remboursé : {df_rembourse['Refunded Amount'].sum():.2f}€")
print(f"Nombre de commandes concernées : {len(df_rembourse)}")
print()
print(df_rembourse[["Name", "Produit", "Total", "Refunded Amount", "Statut simplifie"]])
```

```python
#identifier les produits/marques les plus remboursés
produits_rembourses = df_rembourse.groupby("Produit")["Refunded Amount"].agg(["sum", "count"])
produits_rembourses.columns = ["Montant rembourse", "Nb fois"]
produits_rembourses = produits_rembourses.sort_values("Montant rembourse", ascending=False)

print(produits_rembourses)
```

```python
#focus sur la marque Dometic
df_rembourse_dometic = df_rembourse[df_rembourse["Produit"].str.contains("Dometic", case=False, na=False)]

print("=" * 50)
print("FOCUS SUR LA MARQUE DOMETIC")
print("=" * 50)
print(f"Montant remboursé Dometic : {df_rembourse_dometic['Refunded Amount'].sum():.2f}€")
print(f"Nombre de commandes Dometic remboursées : {len(df_rembourse_dometic)}")
print(f"Part dans le total des remboursements : {df_rembourse_dometic['Refunded Amount'].sum() / df_rembourse['Refunded Amount'].sum() * 100:.1f}%")
```

```python
df_payees = df[df["Statut simplifie"] == "Payee"]
ca_mois = df_payees.groupby("Mois")["Total"].agg(["sum", "count"])
ca_mois.columns = ["CA (€)", "Nb commandes"]

print(ca_mois)
```

```python
#top produits par chiffre d'affaires
top_produits = df_payees.groupby("Produit")["Total"].sum().sort_values(ascending=False).head(10)

print(top_produits)
```

```python
#Statistiques descriptives
print("=" * 50)
print("STATISTIQUES DESCRIPTIVES - COMMANDES PAYÉES")
print("=" * 50)
print(df_payees["Total"].describe())
print()
print(f"Panier moyen : {df_payees['Total'].mean():.2f}€")
print(f"Médiane : {df_payees['Total'].median():.2f}€")
print(f"Écart-type : {df_payees['Total'].std():.2f}€")
print()
print(f"Taux de remboursement : {len(df_rembourse) / len(df) * 100:.1f}%")
```

```python
#Graphique CA par mois (barres)
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(ca_mois.index, ca_mois["CA (€)"], color="steelblue")
ax.set_title("Chiffre d'affaires par mois (commandes payées)")
ax.set_xlabel("Mois")
ax.set_ylabel("CA (€)")
for i, v in enumerate(ca_mois["CA (€)"]):
    ax.text(i, v + 10, f"{v:.0f}€", ha="center", fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
#Camembert des statuts
statuts = df["Statut simplifie"].value_counts()

fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(statuts.values, labels=statuts.index, autopct="%1.1f%%",
       colors=["#4CAF50", "#F44336", "#FF9800"])
ax.set_title("Répartition des commandes par statut")
plt.show()
```

```python
#Top produits (barres horizontales)
fig, ax = plt.subplots(figsize=(10, 6))
top_produits_court = top_produits.copy()
top_produits_court.index = [p[:40] + "..." if len(p) > 40 else p 
                             for p in top_produits_court.index]
ax.barh(top_produits_court.index[::-1], top_produits_court.values[::-1], 
        color="coral")
ax.set_title("Top 10 produits par CA (commandes payées)")
ax.set_xlabel("CA (€)")
plt.tight_layout()
plt.show()
```

```python
#Distribution des montants (histogramme)
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(df_payees["Total"], bins=8, color="mediumpurple", edgecolor="white")
ax.set_title("Distribution des montants des commandes payées")
ax.set_xlabel("Montant (€)")
ax.set_ylabel("Nombre de commandes")
plt.tight_layout()
plt.show()
```
