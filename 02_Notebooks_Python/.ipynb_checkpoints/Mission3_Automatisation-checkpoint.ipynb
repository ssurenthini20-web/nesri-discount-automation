```python
# ============================================================
# MISSION 3 — BRIQUE 1 : Nettoyage du catalogue produits
# Surenthini SIVAKUMAR — Nesri Discount
# ============================================================

import pandas as pd
import re
import os

# ============================================================
# 1. CHARGEMENT DES 5 FICHIERS CSV
# ============================================================

dossier_brut = "/Users/surenthinisivakumar/Documents/Stage_Nesri_Discount/01_Donnees_brutes/"

dfs = []
for i in range(1, 6):
    chemin = os.path.join(dossier_brut, f"products_export_{i}.csv")
    df = pd.read_csv(chemin, low_memory=False)
    print(f"Fichier {i} chargé : {len(df)} lignes")
    dfs.append(df)

df_all = pd.concat(dfs, ignore_index=True)
print(f"\nTotal brut : {len(df_all)} lignes")

# ============================================================
# 2. NETTOYAGE
# ============================================================

# Étape 1 — Garder uniquement les produits actifs
df = df_all[df_all['Status'] == 'active'].copy()
print(f"Après filtre 'active' : {len(df)} lignes")

# Étape 2 — Garder uniquement les produits publiés
df = df[df['Published'] == True].copy()
print(f"Après filtre 'Published' : {len(df)} lignes")

# Étape 3 — Garder uniquement les produits en stock
df = df[pd.to_numeric(df['Variant Inventory Qty'], errors='coerce').fillna(0) > 0].copy()
print(f"Après filtre stock > 0 : {len(df)} lignes")

# Étape 4 — Supprimer les doublons de variantes (1 ligne par produit)
df = df.drop_duplicates(subset='Handle', keep='first').copy()
print(f"Après déduplication : {len(df)} lignes")

# Étape 5 — Nettoyer et préparer les deux prix
df['Prix_vente'] = pd.to_numeric(df['Variant Price'], errors='coerce')
df['Prix_barre'] = pd.to_numeric(df['Variant Compare At Price'], errors='coerce')

# Supprimer les produits sans prix de vente
df = df[df['Prix_vente'] > 0].copy()
print(f"Après filtre prix > 0 : {len(df)} lignes")

# Étape 6 — Calculer l'économie et le % de remise
df['Economie'] = (df['Prix_barre'] - df['Prix_vente']).round(2)
df['Remise_pct'] = ((df['Economie'] / df['Prix_barre']) * 100).round(0)

# Étape 7 — Supprimer les remises aberrantes (erreurs de saisie Shopify)
nb_avant = len(df)
df = df[~(df['Remise_pct'] > 70)].copy()
print(f"Remises aberrantes supprimées : {nb_avant - len(df)} produits")

# Étape 8 — Nettoyer les descriptions HTML
def clean_html(text):
    if pd.isna(text):
        return ""
    text = re.sub(r'<[^>]+>', ' ', str(text))   # supprimer balises HTML
    text = re.sub(r'&[a-z]+;', ' ', text)        # supprimer &nbsp; etc.
    text = re.sub(r'\s+', ' ', text).strip()     # supprimer espaces en trop
    return text[:600]                             # limiter à 600 caractères

df['Description'] = df['Body (HTML)'].apply(clean_html)

# ============================================================
# 3. GARDER UNIQUEMENT LES COLONNES UTILES
# ============================================================

df_final = df[[
    'Handle',
    'Title',
    'Vendor',
    'Prix_vente',
    'Prix_barre',
    'Economie',
    'Remise_pct',
    'Description',
    'Image Src',
    'Marques (product.metafields.custom.marques)',
    'Type',
    'Product Category'
]].copy()

# Renommer les colonnes en français
df_final.columns = [
    'Handle',
    'Titre',
    'Vendeur',
    'Prix_vente',
    'Prix_barre',
    'Economie',
    'Remise_pct',
    'Description',
    'Image',
    'Marque',
    'Type',
    'Categorie'
]

# Réinitialiser l'index pour la rotation
df_final = df_final.reset_index(drop=True)

# ============================================================
# 4. RÉSUMÉ FINAL
# ============================================================

print(f"\n{'='*50}")
print(f"✅ CATALOGUE FINAL : {len(df_final)} produits propres")
print(f"   → {df_final['Prix_barre'].notna().sum()} produits EN PROMO (avec prix barré)")
print(f"   → {df_final['Prix_barre'].isna().sum()} produits à prix fixe")
print(f"   → Remise moyenne : {df_final['Remise_pct'].mean():.0f}%")
print(f"{'='*50}")

# ============================================================
# RÉORGANISATION : produits EN PROMO en premier
# ============================================================

df_promo     = df_final[df_final['Prix_barre'].notna()].copy()
df_prix_fixe = df_final[df_final['Prix_barre'].isna()].copy()

# Trier les promos par économie décroissante (les meilleures deals en premier)
df_promo = df_promo.sort_values('Economie', ascending=False)

# Fusionner : promos d'abord, puis prix fixes
df_final = pd.concat([df_promo, df_prix_fixe], ignore_index=True)

print(f"\n📊 Ordre de rotation :")
print(f"   → Produits 0 à {len(df_promo)-1} : EN PROMO ({len(df_promo)} produits) 🔥")
print(f"   → Produits {len(df_promo)} à {len(df_final)-1} : Prix fixe ({len(df_prix_fixe)} produits)")

# ============================================================
# 5. SAUVEGARDE
# ============================================================

dossier_propre = "/Users/surenthinisivakumar/Documents/Stage_Nesri_Discount/03_Donnees_propres/"

chemin_sortie = os.path.join(dossier_propre, "catalogue_produits_propre.csv")
df_final.to_csv(chemin_sortie, index=False, encoding='utf-8-sig')

print(f"\n💾 Fichier sauvegardé ici :")
print(f"   {chemin_sortie}")
```

```python
# ============================================================
# MISSION 3 — FUSION COMPLET — BRIQUE 3 & BRIQUE 4 (DESIGN GENERATION PREMIUM)
# Surenthini SIVAKUMAR — Nesri Discount
# ============================================================
import os
import smtplib
from email import encoders
from email.mime.base import MIMEBase
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from io import BytesIO
import requests
import pandas as pd
from PIL import Image, ImageDraw, ImageFilter, ImageFont
from dotenv import load_dotenv
from groq import Groq

# ============================================================
# 1. INITIALISATION ET CONFIGURATION
# ============================================================
load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

expediteur = "raadha.2910@gmail.com"
mot_de_passe = "kzfd msgr ykwg duse"
destinataire = "shivadheepi@gmail.com"

# Configuration des dimensions et de la charte graphique premium
CONFIG_IMAGES = {
    "INSTA": {"size": (1080, 1350)},
    "TIKTOK": {"size": (1080, 1920)},
}

COLOR_BG_TOP = (12, 22, 54, 255)       # Bleu nuit moderne pro
COLOR_BG_BOTTOM = (20, 32, 68, 255)
COLOR_TEXT_MAIN = (255, 255, 255, 255) # Blanc pur pour une lisibilité maximale
COLOR_TEXT_MUTED = (170, 185, 220, 255)# Bleu épuré pour la marque
COLOR_ACCENT = (0, 245, 156, 255)      # Vert néon pour le macaron de réduction
COLOR_RED = (227, 6, 19, 255)          # Rouge Nesri officiel pour le bouton et bandeau

# ============================================================
# 2. CHARGEMENT ET SÉLECTION INTELLIGENTE DU PRODUIT (ALGORITHME SAISONNIER)
# ============================================================
dossier_propre = "/Users/surenthinisivakumar/Documents/Stage_Nesri_Discount/03_Donnees_propres/"
df = pd.read_csv(os.path.join(dossier_propre, "catalogue_produits_propre.csv"))

# --- ALGORITHME SAISONNIER D'ÉTÉ (DATA SCIENCE) ---
# 1. On définit les mots-clés des produits recherchés en été (24 juin)
mots_cles_ete = ["frigo", "réfrigérateur", "refrigerateur", "ventilateur", "climatiseur", 
                 "climatisation", "rafraîchisseur", "rafraichisseur", "congélateur", 
                 "congelateur", "glacière", "glaciere", "barbecue", "plancha"]

# 2. On filtre le catalogue pour ne garder QUE les produits d'été (dans le titre, la description ou la catégorie)
masque_ete = df['Titre'].str.lower().str.contains('|'.join(mots_cles_ete)) | \
             df['Description'].str.lower().str.contains('|'.join(mots_cles_ete)) | \
             df['Categorie'].str.lower().str.contains('|'.join(mots_cles_ete))

df_ete = df[masque_ete].copy()

# 3. Parmi ces produits de saison, on prend ceux en promo et on les trie par la meilleure remise
df_promos_ete = df_ete[df_ete['Remise_pct'] > 0].copy()
df_promos_ete = df_promos_ete.sort_values(by=['Remise_pct', 'Economie'], ascending=False)

# Sécurité : Si aucun produit d'été n'est en promo, on prend tous les produits d'été dispo
if df_promos_ete.empty:
    df_promos_ete = df_ete.sort_values(by=['Prix_vente'], ascending=True)

chemin_compteur = os.path.join(dossier_propre, "compteur_rotation_ete.txt")

if not os.path.exists(chemin_compteur):
    index_hier = -1
else:
    with open(chemin_compteur, 'r') as f:
        index_hier = int(f.read().strip())

# Sélection du produit du jour parmi la liste d'été optimisée
index_aujourdhui = (index_hier + 1) % len(df_promos_ete)
produit = df_promos_ete.iloc[index_aujourdhui].copy()

if pd.isna(produit['Marque']) or str(produit['Marque']).strip() == 'nan':
    produit['Marque'] = produit['Vendeur']

# 🛡️ SYSTEME DE SECURITE ET VALIDATION DU PRIX (ANTI-ERREUR)
print("-" * 60)
print("🛡️ SÉCURITÉ CATALOGUE — VÉRIFICATION AVANT GÉNÉRATION")
print("-" * 60)
print(f"📌 Produit d'été du jour : {produit['Titre']}")
print(f"💰 Prix catalogue  : {produit['Prix_vente']} €")
print(f"❌ Prix barre      : {produit['Prix_barre']} €")
print("-" * 60)

confirmation = input("👉 Le prix de vente est-il identique sur le site internet ? (OUI/NON) : ").strip().upper()

if confirmation in ["NON", "N"]:
    print("\n⚠️ Correction du prix demandée...")
    try:
        nouveau_prix = float(input("✍️ Entrez le VRAI PRIX visible sur le site (ex: 24) : "))
        
        # Modification en mémoire pour aujourd'hui
        produit['Prix_vente'] = nouveau_prix
        if pd.notna(produit['Prix_barre']):
            produit['Economie'] = produit['Prix_barre'] - nouveau_prix
            produit['Remise_pct'] = round(((produit['Prix_barre'] - nouveau_prix) / produit['Prix_barre']) * 100)
        
        # Correction dans le gros fichier d'origine
        idx_original = produit.name
        df.loc[idx_original, 'Prix_vente'] = nouveau_prix
        if pd.notna(df.loc[idx_original, 'Prix_barre']):
            df.loc[idx_original, 'Economie'] = df.loc[idx_original, 'Prix_barre'] - nouveau_prix
            df.loc[idx_original, 'Remise_pct'] = round(((df.loc[idx_original, 'Prix_barre'] - nouveau_prix) / df.loc[idx_original, 'Prix_barre']) * 100)
        
        df.to_csv(os.path.join(dossier_propre, "catalogue_produits_propre.csv"), index=False, encoding='utf-8-sig')
        print("💾 Le fichier catalogue_produits_propre.csv a été mis à jour avec le nouveau prix !")
        
    except ValueError:
        print("🚨 Saisie invalide. On conserve le prix d'origine.")
else:
    print("\n✅ Prix validé ! Lancement de la création...")

# Formatage des affichages
def format_prix(prix):
    if pd.isna(prix):
        return ""
    return f"{int(prix):,}€".replace(",", " ")
                                              
produit['Prix_vente_affiche'] = format_prix(produit['Prix_vente'])
produit['Prix_barre_affiche'] = format_prix(produit['Prix_barre'])
produit['Economie_affiche']   = format_prix(produit['Economie'])


# ============================================================
# 3. GÉNÉRATION DU POST INSTAGRAM ADAPTÉE À LA CANICULE (NLP)
# ============================================================
categorie_produit = str(produit.get('Categorie', '')).lower()
titre_produit = str(produit['Titre']).lower()

# Instructions par défaut
consigne_style = "Adopte un ton dynamique et estival."

# NLP ultra-ciblé sur la fraîcheur et la chaleur de l'été
if any(x in titre_produit or x in categorie_produit for x in ["froid", "refrigerateur", "frigo", "congelateur", "congélateur"]):
    consigne_style = "Mets l'accent sur la fraîcheur indispensable en plein été ! Parle de boissons bien fraîches, de glaces et de la grande capacité pour stocker les grillades du barbecue."
elif any(x in titre_produit or x in categorie_produit for x in ["ventilateur", "climatiseur", "rafraichisseur", "rafraîchisseur"]):
    consigne_style = "Insiste sur la canicule et la chaleur actuelle. Propose ce produit comme LA solution pour passer des nuits au frais et respirer chez soi."
elif "barbecue" in categorie_produit or "camping" in categorie_produit:
    consigne_style = "Mets l'accent sur les longues soirées d'été en terrasse, la convivialité, les repas en extérieur et le partage en famille."

if pd.notna(produit['Prix_barre']):
    info_prix = (
        f"Prix original : {produit['Prix_barre_affiche']}\n"
        f"Prix actuel : {produit['Prix_vente_affiche']}\n"
        f"Économie : {produit['Economie_affiche']} soit -{int(produit['Remise_pct'])}% de remise"
    )
else:
    info_prix = f"Prix : {produit['Prix_vente_affiche']}"

prompt = f"""Tu es un expert en marketing digital pour la boutique Nesri Discount.
Génère un post Instagram hyper accrocheur pour ce produit phare de l'été :
Nom du produit : {produit['Titre']}
Marque : {produit['Marque']}
{info_prix}
Description : {produit['Description'][:400]}

Règle de style contextuelle (Saison ÉTÉ) :
- {consigne_style}

Règles générales :
- Accroche percutante axée sur l'été/la chaleur avec des emojis adaptés (☀️, 🧊, 🥵, 🌊)
- Mets bien en valeur la promo discount
- Maximum 8 hashtags pertinents sans accent (ex: #frais #canicule #ete #electro)
- Entre 6 et 10 lignes max avec des sauts de ligne pour que ce soit aéré
- Termine par : 🛒 Disponible sur NesriDiscount.fr
"""

print(f"🤖 IA : Génération du post de saison pour {produit['Titre'][:40]}...")
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": prompt}],
    max_tokens=500,
    temperature=0.8
)
post_genere = response.choices[0].message.content

# ============================================================
# 4. FONCTIONS OUTILS DU RENDU VISUEL (PIL)
# ============================================================
def load_font(size, bold=False):
    candidates = [
        "/Library/Fonts/Arial Bold.ttf" if bold else "/Library/Fonts/Arial.ttf",
        "/System/Library/Fonts/Supplemental/Arial Bold.ttf" if bold else "/System/Library/Fonts/Supplemental/Arial.ttf",
    ]
    for path in candidates:
        if os.path.exists(path):
            return ImageFont.truetype(path, size)
    return ImageFont.load_default()

def text_width(draw, text, font):
    bbox = draw.textbbox((0, 0), text, font=font)
    return bbox[2] - bbox[0]

def fit_font(draw, text, max_width, start=42, min_size=28, bold=True):
    for size in range(start, min_size - 1, -2):
        font = load_font(size, bold=bold)
        if text_width(draw, text, font) <= max_width:
            return font
    return load_font(min_size, bold=bold)

def wrap_text(draw, text, font, max_width, max_lines=2):
    words = text.split()
    lines = []
    current = ""
    for w in words:
        test = w if not current else current + " " + w
        if text_width(draw, test, font) <= max_width:
            current = test
        else:
            if current:
                lines.append(current)
            current = w
    if current:
        lines.append(current)
    return "\n".join(lines[:max_lines])

def safe_int(v, default=0):
    try:
        if v is None or pd.isna(v):
            return default
        return int(float(v))
    except:
        return default

# ============================================================
# 5. MOTEUR GRAPHIQUE SUBLIMÉ (LOGO AGRANDI + EFFET GLOW ATTRACTIF)
# ============================================================
def generate_creative_in_memory(format_type, produit_data):
    W, H = CONFIG_IMAGES[format_type]["size"]
    canvas = Image.new("RGBA", (W, H))
    draw = ImageDraw.Draw(canvas)

    # 1. Fond dégradé linéaire haut de gamme
    for y in range(H):
        t = y / H
        r = int(COLOR_BG_TOP[0] + (COLOR_BG_BOTTOM[0] - COLOR_BG_TOP[0]) * t)
        g = int(COLOR_BG_TOP[1] + (COLOR_BG_BOTTOM[1] - COLOR_BG_TOP[1]) * t)
        b = int(COLOR_BG_TOP[2] + (COLOR_BG_BOTTOM[2] - COLOR_BG_TOP[2]) * t)
        draw.line([(0, y), (W, y)], fill=(r, g, b, 255))

    # 2. Halos de lumière d'ambiance d'arrière-plan
    glow = Image.new("RGBA", (W, H), (0, 0, 0, 0))
    gd = ImageDraw.Draw(glow)
    gd.ellipse((W//2 - 350, H//2 - 350, W//2 + 350, H//2 + 350), fill=(255, 255, 255, 10))
    gd.ellipse((W - 300, -100, W + 300, 500), fill=(COLOR_ACCENT[0], COLOR_ACCENT[1], COLOR_ACCENT[2], 12))
    gd.ellipse((-200, H - 600, 400, H + 200), fill=(COLOR_RED[0], COLOR_RED[1], COLOR_RED[2], 12))
    glow = glow.filter(ImageFilter.GaussianBlur(130))
    canvas = Image.alpha_composite(canvas, glow)
    draw = ImageDraw.Draw(canvas)

    # 3. Fine bordure élégante
    draw.rectangle([40, 40, W - 40, H - 40], outline=(255, 255, 255, 15), width=2)

    # 4. Bandeau "OFFRE FLASH" en haut à droite
    bandeau = Image.new("RGBA", (W, H), (0, 0, 0, 0))
    bd = ImageDraw.Draw(bandeau)
    bd.polygon([(W - 350, 0), (W, 350), (W, 230), (W - 230, 0)], fill=COLOR_RED)
    canvas = Image.alpha_composite(canvas, bandeau)
    draw = ImageDraw.Draw(canvas)
    
    f_flash = load_font(24, bold=True)
    draw.text((W - 195, 75), "OFFRE", font=f_flash, fill=(255, 255, 255, 255))
    draw.text((W - 185, 110), "FLASH", font=f_flash, fill=(255, 255, 255, 255))

    # 5. INTÉGRATION LOGO : AGRANDI ET EMBELLI AVEC EFFET GLOW PREMIUM
    logo_path = "/Users/surenthinisivakumar/Documents/Stage_Nesri_Discount/logo_nesri.png"
    if os.path.exists(logo_path):
        try:
            logo = Image.open(logo_path).convert("RGBA")
            logo_w = 380  # Super grand et bien visible pour marquer l'esprit des clients !
            logo_h = int(logo.height * (logo_w / logo.width))
            logo = logo.resize((logo_w, logo_h))
            
            # Création d'un halo lumineux blanc/bleuté ultra-tendance juste derrière le logo
            logo_glow = Image.new("RGBA", (W, H), (0, 0, 0, 0))
            lg_draw = ImageDraw.Draw(logo_glow)
            lg_draw.ellipse([30, 30, 30 + logo_w + 60, 60 + logo_h + 30], fill=(255, 255, 255, 25))
            logo_glow = logo_glow.filter(ImageFilter.GaussianBlur(25))
            
            # Fusion du halo et du logo sur l'image principale
            canvas.alpha_composite(logo_glow, (0, 0))
            canvas.alpha_composite(logo, (60, 60))
        except:
            pass

    # 6. Image Produit Centrale
    py_prod = int(H * 0.25) if format_type == "TIKTOK" else int(H * 0.20)
    h_prod_box = int(W * 0.60)

    image_url = produit_data.get("Image")
    if pd.notna(image_url) and str(image_url).strip() != "":
        try:
            res = requests.get(image_url, timeout=15)
            img_prod = Image.open(BytesIO(res.content)).convert("RGBA")
            img_prod.thumbnail((h_prod_box, h_prod_box))
            px = (W - img_prod.width) // 2

            shadow = Image.new("RGBA", (img_prod.width + 100, img_prod.height + 100), (0, 0, 0, 0))
            sd = ImageDraw.Draw(shadow)
            sd.ellipse((50, img_prod.height + 5, img_prod.width + 50, img_prod.height + 25), fill=(0, 0, 0, 120))
            shadow = shadow.filter(ImageFilter.GaussianBlur(15))
            canvas.alpha_composite(shadow, (px - 50, py_prod - 50))
            canvas.alpha_composite(img_prod, (px, py_prod))
        except Exception as e:
            print(f"Erreur image produit : {e}")

    # 7. Textes Marketing (Marque + Titre)
    text_y_start = int(H * 0.67) if format_type == "TIKTOK" else int(H * 0.60)
    
    marque_txt = str(produit_data.get("Marque", "EXCLUSIVITÉ")).upper()
    if marque_txt == "NAN":
        marque_txt = "NESRI DISCOUNT"
    
    draw.text((60, text_y_start), marque_txt, font=load_font(24, bold=True), fill=COLOR_TEXT_MUTED)

    titre_full = str(produit_data.get("Titre", "Produit"))
    title_font = fit_font(draw, titre_full, W - 120, start=42, min_size=28, bold=True)
    title_wrapped = wrap_text(draw, titre_full, title_font, W - 120)
    
    draw.multiline_text((60, text_y_start + 45), title_wrapped, font=title_font, fill=COLOR_TEXT_MAIN, spacing=6)
    
    # 8. SECTION PRIX HORIZONTALE ET PROPRE
    nb_lignes_titre = len(title_wrapped.split('\n'))
    hauteur_par_ligne = title_font.getbbox("A")[3] - title_font.getbbox("A")[1] + 6
    prix_y = text_y_start + 45 + (nb_lignes_titre * hauteur_par_ligne) + 30

    prix_vente = safe_int(produit_data.get("Prix_vente"), 0)
    prix_barre = safe_int(produit_data.get("Prix_barre"), 0)
    remise = safe_int(produit_data.get("Remise_pct"), 0)

    # Affichage du Prix de Vente Principal
    txt_pv = f"{prix_vente} €"
    draw.text((60, prix_y), txt_pv, font=load_font(85, bold=True), fill=COLOR_TEXT_MAIN)

    if prix_barre > 0 and remise > 0:
        largeur_pv = text_width(draw, txt_pv, load_font(85, bold=True))
        next_x = 60 + largeur_pv + 35
        
        # Affichage du prix d'origine
        txt_pb = f"{prix_barre} €"
        draw.text((next_x, prix_y + 35), txt_pb, font=load_font(36), fill=COLOR_TEXT_MUTED)
        
        # Ligne rouge très fine de barrage
        largeur_pb = text_width(draw, txt_pb, load_font(36))
        draw.line([(next_x, prix_y + 53), (next_x + largeur_pb, prix_y + 53)], fill=COLOR_RED, width=3)

        # Macaron vert néon de réduction à droite
        draw.rounded_rectangle([W - 240, prix_y + 10, W - 60, prix_y + 75], radius=14, fill=COLOR_ACCENT)
        txt_remise = f"-{remise}%"
        tw_rem = text_width(draw, txt_remise, load_font(32, bold=True))
        draw.text((W - 150 - (tw_rem // 2), prix_y + 22), txt_remise, font=load_font(32, bold=True), fill=(12, 22, 54, 255))

    # 9. Bouton CTA stabilisé en bas
    draw.rounded_rectangle([60, H - 150, W - 60, H - 60], radius=22, fill=COLOR_RED)
    cta_txt = "PROFITER DE L'OFFRE SUR LE SITE  ➔"
    tw_cta = text_width(draw, cta_txt, load_font(28, bold=True))
    draw.text(((W - tw_cta) // 2, H - 118), cta_txt, font=load_font(28, bold=True), fill=(255, 255, 255, 255))

    img_byte_arr = BytesIO()
    canvas.convert("RGB").save(img_byte_arr, format="PNG", optimize=True)
    img_byte_arr.seek(0)
    return img_byte_arr.read()

# ============================================================
# 6. CONSTRUIRE LE MAIL HTML ET ATTACHER LES IMAGES
# ============================================================
if pd.notna(produit['Prix_barre']):
    bloc_prix = f"""
    <p style="font-size:14px;color:#888;margin:0 0 4px;"><s>{produit['Prix_barre_affiche']}</s></p>
    <span style="font-size:36px;font-weight:700;color:#E30613;">{produit['Prix_vente_affiche']}</span>
    <span style="display:inline-block;background:#E30613;color:#fff;font-size:13px;font-weight:700;padding:4px 10px;border-radius:6px;margin-left:10px;">-{int(produit['Remise_pct'])}%</span>
    <p style="font-size:14px;color:#2E7D32;font-weight:700;margin:6px 0 0;">Vous économisez {produit['Economie_affiche']} !</p>
    """
else:
    bloc_prix = f"""<span style="font-size:36px;font-weight:700;color:#E30613;">{produit['Prix_vente_affiche']}</span>"""

html = f"""
<!DOCTYPE html>
<html>
<body style="margin:0;padding:0;background:#f4f4f4;font-family:Arial,sans-serif;">
<div style="max-width:580px;margin:20px auto;background:#ffffff;border-radius:12px;overflow:hidden;">
  <div style="background:#000000;padding:24px 32px;text-align:center;">
    <p style="color:#fff;font-size:24px;font-weight:700;margin:0 0 4px;letter-spacing:1px;">NESRI DISCOUNT</p>
    <p style="color:#ffccce;font-size:13px;margin:0;">Électroménager discount depuis 1956 - Paris</p>
  </div>
  <img src="{produit['Image']}" alt="{produit['Titre']}" style="width:100%;height:320px;object-fit:contain;background:#f8f8f8;display:block;" />
  <div style="padding:24px 32px;">
    <p style="font-size:20px;font-weight:700;color:#1a1a1a;margin:0 0 4px;">{produit['Titre']}</p>
    <p style="font-size:13px;color:#888;margin:0 0 16px;text-transform:uppercase;letter-spacing:1px;">{produit['Marque']}</p>
    <div style="margin-bottom:20px;">{bloc_prix}</div>
    <hr style="border:none;border-top:1px solid #eee;margin:0 0 20px;" />
    <p style="font-size:15px;color:#333;line-height:1.7;background:#fafafa;border-left:3px solid #E30613;padding:14px 16px;border-radius:8px;margin:0 0 24px;white-space:pre-wrap;">{post_genere}</p>
  </div>
</div>
</body>
</html>
"""

message = MIMEMultipart("mixed")
message["From"] = expediteur
message["To"] = destinataire
message["Subject"] = f"Offre du jour IA - Nesri Discount ({produit['Titre'][:30]})"

msg_alternative = MIMEMultipart("alternative")
msg_alternative.attach(MIMEText(html, "html"))
message.attach(msg_alternative)

print("🎨 Création des visuels haut de gamme...")
data_insta = generate_creative_in_memory("INSTA", produit)
data_tiktok = generate_creative_in_memory("TIKTOK", produit)

part_insta = MIMEBase("application", "octet-stream")
part_insta.set_payload(data_insta)
encoders.encode_base64(part_insta)
part_insta.add_header("Content-Disposition", 'attachment; filename="pub_instagram.png"')
message.attach(part_insta)

part_tiktok = MIMEBase("application", "octet-stream")
part_tiktok.set_payload(data_tiktok)
encoders.encode_base64(part_tiktok)
part_tiktok.add_header("Content-Disposition", 'attachment; filename="pub_tiktok.png"')
message.attach(part_tiktok)

try:
    print("🚀 Envoi du mail en cours...")
    with smtplib.SMTP_SSL("smtp.gmail.com", 465) as serveur:
        serveur.login(expediteur, mot_de_passe)
        serveur.sendmail(expediteur, destinataire, message.as_string())
    print("🎉 Magnifique ! Le logo est ultra-attractif avec son effet de lumière.")
except Exception as e:
    print(f"Erreur d'envoi : {e}")

# ============================================================
# 7. METTRE À JOUR LE COMPTEUR DE ROTATION
# ============================================================
with open(chemin_compteur, 'w') as f:
    f.write(str(index_aujourdhui))
print(f"✅ Compteur mis à jour → produit suivant enregistré.")
```
